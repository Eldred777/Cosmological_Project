# 3. Fitting models to data

In this workbook we will import data and find the cosmological model that best fits that data. 

### 3.1 First set up the packages and functions
We need to know the distance modulus as a function of redshift.  Distance modulus is defined as 
\begin{equation} \mu(z) = 5 \log_{10}(D_L)\end{equation}
where
\begin{equation} D_L = R_0 S_k(\chi) (1+z) \end{equation}
and $S_k(\chi)=\sin(\chi), \chi,\; {\rm or}\; \sinh(\chi)$ for closed, flat, and open universes respectively, and
\begin{equation} \chi = \frac{c}{R_0H_0}\int_0^z \frac{dz}{E(z)}\end{equation}
with $E(z)=H(z)/H_0$.

Note that the absolute magnitude of the SNe is about -19.5, but it enters as an additive constant in the same way as $c/H_0$ and both have high uncertainty.  Therefore we *marginalise* over this offset and it doesn't matter what value of $H_0$ we use.

```python
# First let's set up our packages
import numpy as np
import math
import random
import threading
from matplotlib import pyplot as plt
from scipy import integrate

# And set some constants
c = 299792.458 # km/s (speed of light)
H0kmsmpc = 70.0  # Hubble constant in km/s/Mpc
H0s = H0kmsmpc * 3.2408e-20 # H0 in inverse seconds is H0 in km/s/Mpc * (3.2408e-20 Mpc/km)

# Write a function for the integrand, i.e. $1/E(z)$,
def ezinv(z,om=0.3,ol=0.7,w0=-1.0,wa=0.0,orr=0.0):
    ok = 1.0-om-ol-orr
    wm = 0
    wr = 1.0/3.0
    wk = -1.0/3.0
    ez = (orr*((1+z)**(3*(1+wr))) + om*((1+z)**(3*(1+wm))) + ol*((1+z)**(3*(1+w0+wa)))*math.exp(-3*wa*(1-(1/(1+z)))) + ok*((1+z)**(3*(1+wk))))**(1/2)
    return 1.0/ez

# The curvature correction function
def Sk(xx, ok):
    if ok < 0.0:
        dk = np.sin(np.sqrt(-ok)*xx)/np.sqrt(-ok)
    elif ok > 0.0:
        dk = np.sinh(np.sqrt(ok)*xx)/np.sqrt(ok)
    else:
        dk = xx
    return dk

# The distance modulus
def dist_mod(zs,om=0.3,ol=0.7,w0=-1.0,wa=0.0,orr=0.0):
    """ Calculate the distance modulus, correcting for curvature"""
    ok = 1.0 - om - ol
    xx = np.array([integrate.quad(ezinv, 0, z, args=(om, ol, w0, wa, orr))[0] for z in zs])
    D = Sk(xx, ok)
    lum_dist = D * (1 + zs) 
    dist_mod = 5 * np.log10(lum_dist) # Distance modulus
    # Add an arbitrary constant that's approximately the log of c on Hubble constant minus absolute magnitude of -19.5
    dist_mod = dist_mod + np.log(c/H0kmsmpc)-(-19.5)  # You can actually skip this step and it won't make a difference to our fitting
    return dist_mod

    
```

### 3.4 Normalise the theory
You can see that there is a vertical offset between the different theory curves.  
We marginalise over this vertical offset, because we do not have enough information to constrain the vertical offset independently.
In other words we use the data to determine what the vertical offset should be, but we do not care about what the answer is!
(We are interested in the other parameters in the fit, but not that one.  The other parameters determine the *shape* of the curve.)

The easiest way to normalise the curves is to take a weighted average of the difference between the data points and the theory in each case.

```python
# Calculate mscript for each of these, which is the thing that determines the vertical normalisation 
mscr_om10_ox00 = np.sum((mu_om10_ox00-mu)/muerr**2)/np.sum(1./muerr**2)
mscr_om03_ox00 = np.sum((mu_om03_ox00-mu)/muerr**2)/np.sum(1./muerr**2)
mscr_om03_ox07 = np.sum((mu_om03_ox07-mu)/muerr**2)/np.sum(1./muerr**2)

print(mscr_om10_ox00,mscr_om03_ox00,mscr_om03_ox07)

mu_om10_ox00=mu_om10_ox00-mscr_om10_ox00
mu_om03_ox00=mu_om03_ox00-mscr_om03_ox00
mu_om03_ox07=mu_om03_ox07-mscr_om03_ox07

# Repeat the plot and see how it changes
# Plot it to see what it looks like, this is called a Hubble diagram
plt.errorbar(zs,mu,yerr=muerr,fmt='.',elinewidth=0.7,markersize=4,alpha=0.5)
plt.plot(zs,mu_om10_ox00,'-.',color='red',label='(1.0, 0.0)')
plt.plot(zs,mu_om03_ox00,'--',color='blue',label='(0.3, 0.0)')
plt.plot(zs,mu_om03_ox07,'-',color='green',label='(0.3, 0.7)')
plt.xlim(0,1.0)
plt.xlabel('Redshift')
plt.ylabel('Distance Modulus')
plt.legend(frameon=False)
plt.show()

# Now plot a Hubble diagram relative to the empty model (i.e. subtract the empty model from all the data and models)
plt.errorbar(zs,mu-mu_om00_ox00,yerr=muerr,fmt='.',elinewidth=0.7,markersize=4,alpha=0.5)
plt.plot(zs,mu_om10_ox00-mu_om00_ox00,'-.',color='red',label='(1.0, 0.0)')
plt.plot(zs,mu_om03_ox00-mu_om00_ox00,'--',color='blue',label='(0.3, 0.0)')
plt.plot(zs,mu_om03_ox07-mu_om00_ox00,'-',color='green',label='(0.3, 0.7)')
plt.axhline(y=0.0,ls=':',color='black')
plt.xlim(0,1.0)
plt.xlabel('Redshift')
plt.ylabel('Distance Modulus Normalised to (0,0)')
plt.legend(frameon=False)
plt.show()
```

## 3.5 Perform a fit to many models 
To find the best model that best matches the data use a $\chi^2$ test.


```python
# Set up the arrays for the models you want to test, e.g. a range of Omega_m and Omega_Lambda models:
n = 21                          # Increase this for a finer grid
oms = np.linspace(0.0, 0.5, n)   # Array of matter densities
ols = np.linspace(0.0, 1.0, n)   # Array of cosmological constant values
chi2 = np.ones((n, n)) * np.inf  # Array to hold our chi2 values, set initially to super large values

# Calculate Chi2 for each model
for i, om in enumerate(oms):                                          # loop through matter densities
        for j, ol in enumerate(ols):                                  # loop through cosmological constant densities
            mu_model = dist_mod(zs, om=om, ol=ol)                     # calculate the distance modulus vs redshift for that model 
            mscr = np.sum((mu_model-mu)/muerr**2)/np.sum(1./muerr**2) # Calculate the vertical offset to apply
            mu_model_norm = mu_model-mscr                             # Apply the vertical offset
            chi2[i,j] = np.sum((mu_model_norm - mu) ** 2 / muerr**2)  # Calculate the chi2 and save it in a matrix
            
# Convert that to a likelihood and calculate the reduced chi2
likelihood = np.exp(-0.5 * (chi2-np.amin(chi2)))  # convert the chi^2 to a likelihood (np.amin(chi2) calculates the minimum of the chi^2 array)
chi2_reduced = chi2 / (len(mu)-2)                 # calculate the reduced chi^2, i.e. chi^2 per degree of freedom, where dof = number of data points minus number of parameters being fitted 

# Calculate the best fit values (where chi2 is minimum)
indbest = np.argmin(chi2)                 # Gives index of best fit but where the indices are just a single number
ibest   = np.unravel_index(indbest,[n,n]) # Converts the best fit index to the 2d version (i,j)
print( 'Best fit values are (om,ol)=(%.3f,%.3f)'%( oms[ibest[0]], ols[ibest[1]] ) )
print( 'Reduced chi^2 for the best fit is %0.2f'%chi2_reduced[ibest[0],ibest[1]] )
```

```python
# Plot contours of 1, 2, and 3 sigma
plt.contour(oms,ols,np.transpose(chi2-np.amin(chi2)),cmap="winter",**{'levels':[2.30,6.18,11.83]})
plt.plot(oms[ibest[0]], ols[ibest[1]],'x',color='black',label='(om,ol)=(%.3f,%.3f)'%( oms[ibest[0]], ols[ibest[1]]) )
plt.xlabel("$\Omega_m$", fontsize=12)
plt.ylabel("$\Omega_\Lambda$", fontsize=12)
plt.plot([oms[0],oms[1]], [ols[0],ols[1]],'-',color='black',label='Step size indicator' ) # Delete this line after making step size smaller!
plt.legend(frameon=False)
plt.savefig('C:/Users/danwe/Cosmology Project/Plots/contours.png', bbox_inches="tight", transparent=True)
plt.show()
plt.close()
```

```python
# THIS IS FOR FITTING 2 PARAMETERS (om,ol)
# Generate some parameter arrays to store all models with 0.85 < reduced chi2 < 1.15
chi2array = [0]
omarray = [0]
olarray = [0]

likelihoodarray = [0]
omlikelihoodarray = [0]
ollikelihoodarray = [0]

# Generate an initial model
omsaved = random.uniform(0,1)
olsaved = random.uniform(0,1)

# Generate priors
omprior = 0.3
delta_omprior = 0.03
prior_om = ((omsaved-omprior)/(delta_omprior))**2

olprior = 0.7
delta_olprior = 0.05
prior_ol = ((olsaved-olprior)/(delta_olprior))**2

# Generate an initial model
mu_model = dist_mod(zs, om=omsaved, ol=olsaved)                                                              
mscr = np.sum((mu_model-mu)/muerr**2)/np.sum(1./muerr**2)                                                    
mu_model_norm = mu_model-mscr                                                                               
chi2I = np.sum((mu_model_norm - mu) ** 2 / muerr**2) #+ prior_om + prior_ol

likelihood = np.exp(-chi2I/2)
likelihoodarray.append(likelihood)
omlikelihoodarray.append(omsaved)
ollikelihoodarray.append(olsaved)

# Define number of Monte Carlo steps, burn in time, and Markov Chain step distance
n = 5000
burntime = 2000
step = 0.05

# Initiate MCMC
for i in range(n):
    
    # Generate new random fit parameters around the old parameters
    om = random.uniform(omsaved-step,omsaved+step)
    ol = random.uniform(olsaved-step,olsaved+step)
    
    # Checking if any values exceed their bounds
    if om < 0:
        om = 0
    if om > 1:
        om = 1
    
    if ol < 0:
        ol = 0
    if ol > 1:
        ol = 1
        
    # Updating priors
    prior_om = ((om-omprior)/(delta_omprior))**2
    prior_ol = ((ol-olprior)/(delta_olprior))**2
  
    # Determine the properties of this new model
    mu_model = dist_mod(zs, om=om, ol=ol)                          
    mscr = np.sum((mu_model-mu)/muerr**2)/np.sum(1./muerr**2)     
    mu_model_norm = mu_model-mscr                                  
    chi2F = np.sum((mu_model_norm - mu) ** 2 / muerr**2) #+ prior_om + prior_ol
    likelihood = np.exp(-chi2F/2)
    likelihoodarray.append(likelihood)
    omlikelihoodarray.append(om)
    ollikelihoodarray.append(ol)
    
    # Determine the ratio of likelihoods
    ratio = math.exp(0.5*(chi2I-chi2F))                            # Determine the ratio of likelihood's between this model and the previous
    ratioTest = random.uniform(0,1)                                # Generate a random number to compare with the ratio
    
    # If the random number is less than the ratio (which it will always be for likely models, and sometimes may be for unlikely)
    if ratioTest < ratio:
        
        # Update the model to be this new model, and update the saved best fit parameters
        chi2I = chi2F
        chi2_reduced = chi2F / (len(mu)-2)
        omsaved = om
        olsaved = ol
        
        if i > burntime:
            chi2array.append(chi2_reduced)
            omarray.append(omsaved)
            olarray.append(olsaved)

chi2array.pop(0)
omarray.pop(0)
olarray.pop(0)

likelihoodarray.pop(0)
omlikelihoodarray.pop(0)
ollikelihoodarray.pop(0)

delta_om = 0.5*(max(omarray)-min(omarray))
delta_ol = 0.5*(max(olarray)-min(olarray))
omsaved = sum(omarray)/len(omarray)
olsaved = sum(olarray)/len(olarray)
chi2_reduced = sum(chi2array)/len(chi2array)
            
print('Best fit values are (om,ol)=(%.3f \u00B1 %.3f, %.3f \u00B1 %.3f'%(omsaved, delta_om, olsaved, delta_ol))
print('Reduced chi^2 for the best fit is %0.2f'%chi2_reduced)

# Generate the final model
emptymodel = dist_mod(zs,om=0.0,ol=0.0)
model = dist_mod(zs,om=omsaved,ol=olsaved)
modelscr = np.sum((model-mu)/muerr**2)/np.sum(1./muerr**2)
model = model-modelscr

modelMAX = dist_mod(zs,om=omsaved+delta_om,ol=olsaved+delta_ol)
modelscrMAX = np.sum((model-mu)/muerr**2)/np.sum(1./muerr**2)
modelMAX = modelMAX-modelscrMAX

modelMIN = dist_mod(zs,om=omsaved-delta_om,ol=olsaved-delta_ol)
modelscrMIN = np.sum((model-mu)/muerr**2)/np.sum(1./muerr**2)
modelMIN = modelMIN-modelscrMIN

# Plot it to see what it looks like, this is called a Hubble diagram
plt.errorbar(zs,mu,yerr=muerr,fmt='.',elinewidth=0.7,markersize=4,alpha=0.5)
plt.plot(zs,model,'-.',color='red',label='Model')
plt.plot(zs,modelMAX,'-.',color='black',label='Upper Error')
plt.plot(zs,modelMIN,'-.',color='black',label='Lower Error')
plt.xlim(0,1.0)
plt.xlabel('Redshift')
plt.ylabel('Distance Modulus')
plt.legend(frameon=False)
plt.show()

# Now plot a Hubble diagram relative to the empty model (i.e. subtract the empty model from all the data and models)
plt.errorbar(zs,mu-emptymodel,yerr=muerr,fmt='.',elinewidth=0.7,markersize=4,alpha=0.5)
plt.plot(zs,model-emptymodel,'-.',color='red',label='Model')
plt.plot(zs,modelMAX-emptymodel,'-.',color='black',label='Upper Error')
plt.plot(zs,modelMIN-emptymodel,'-.',color='black',label='Lower Error')
plt.axhline(y=0.0,ls=':',color='black')
plt.xlim(0,1.0)
plt.xlabel('Redshift')
plt.ylabel('Distance Modulus Normalised to (0,0)')
plt.legend(frameon=False)
plt.show()

fig, axs = plt.subplots(2)
fig.suptitle('Likelihoods')
axs[0].plot(omlikelihoodarray, likelihoodarray, '.')
axs[1].plot(ollikelihoodarray, likelihoodarray, '.')
```

```python
# THIS IS FOR FITTING 5 PARAMETERS
# Generate some parameter arrays to store all models with 0.85 < reduced chi2 < 1.15 
omarray = [0]
olarray = [0]
orrarray = [0]
w0array = [0]
waarray = [0]

likelihoodarray = [0]
omlikelihoodarray = [0]
ollikelihoodarray = [0]
orrlikelihoodarray = [0]
w0likelihoodarray = [0]
walikelihoodarray = [0]

# Generate an initial model
omsaved = random.uniform(0,1)
olsaved = random.uniform(0,1)
orrsaved = random.uniform(0,1)
w0saved = random.uniform(-2,0)
wasaved = random.uniform(-1.5,0.5)

# Generate priors
omprior = 0.3
delta_omprior = 0.03
prior_om = ((omsaved-omprior)/(delta_omprior))**2

olprior = 0.5
delta_olprior = 0.05
prior_ol = ((olsaved-olprior)/(delta_olprior))**2

orrprior = 0
delta_orrprior = 1
prior_orr = ((orrsaved-orrprior)/(delta_orrprior))**2

w0prior = -1
delta_w0prior = 1
prior_w0 = ((w0saved-w0prior)/(delta_w0prior))**2

waprior = 0
delta_waprior = 1
prior_wa = ((wasaved-waprior)/(delta_waprior))**2

# Generate an initial model
mu_model = dist_mod(zs, om=omsaved, ol=olsaved, w0=w0saved, wa=wasaved, orr=orrsaved)                          # Calculate the distance modulus vs redshift for that model 
mscr = np.sum((mu_model-mu)/muerr**2)/np.sum(1./muerr**2)                                                      # Calculate the vertical offset to apply
mu_model_norm = mu_model-mscr                                                                                  # Apply the vertical offset
chi2I = np.sum((mu_model_norm - mu) ** 2 / muerr**2) #+ prior_om + prior_ol + prior_orr + prior_w0 + prior_wa  # Calculate the chi2

likelihood = np.exp(-chi2I/2)
likelihoodarray.append(likelihood)
omlikelihoodarray.append(omsaved)
ollikelihoodarray.append(olsaved)
orrlikelihoodarray.append(orrsaved)
w0likelihoodarray.append(w0saved)
walikelihoodarray.append(wasaved)

# Define number of Monte Carlo steps and Markov Chain step distance
n = 5000
step = 0.1

# Initiate MCMC
for i in range(n):
    
    # Generate new random fit parameters around the old parameters
    om = random.uniform(omsaved-step,omsaved+step)
    ol = random.uniform(olsaved-step,olsaved+step)
    orr = random.uniform(orrsaved-step,orrsaved+step)
    w0 = random.uniform(w0saved-step,w0saved+step)
    wa = random.uniform(wasaved-step,wasaved+step)
    
    # Checking if any values exceed their bounds
    if om < 0:
        om = 0
    if om > 1:
        om = 1
    
    if ol < 0:
        ol = 0
    if ol > 1:
        ol = 1
    
    if orr < 0:
        orr = 0
    if orr > 1:
        orr = 1

    if w0 < -2:
        w0 = -2
    if w0 > 0:
        w0 = 0

    if wa < -1.5:
        wa = -1.5
    if wa > 0.5:
        wa = 0.5

    # Update priors
    prior_om = ((om-omprior)/(delta_omprior))**2
    prior_ol = ((ol-olprior)/(delta_olprior))**2
    prior_orr = ((orr-orrprior)/(delta_orrprior))**2
    prior_w0 = ((w0-w0prior)/(delta_w0prior))**2
    prior_wa = ((wa-waprior)/(delta_waprior))**2    
  
    # Determine the properties of this new model
    mu_model = dist_mod(zs, om=om, ol=ol, w0=w0, wa=wa, orr=orr) 
    mscr = np.sum((mu_model-mu)/muerr**2)/np.sum(1./muerr**2)
    mu_model_norm = mu_model-mscr                                 
    chi2F = np.sum((mu_model_norm - mu) ** 2 / muerr**2) #+ prior_om + prior_ol + prior_orr + prior_w0 + prior_wa
    
    likelihood = np.exp(-chi2F/2)
    likelihoodarray.append(likelihood)
    omlikelihoodarray.append(om)
    ollikelihoodarray.append(ol)
    orrlikelihoodarray.append(orr)
    w0likelihoodarray.append(w0)
    walikelihoodarray.append(wa)
    
    # Determine the ratio of likelihoods
    ratio = math.exp(0.5*(chi2I-chi2F))                            # Determine the ratio of likelihood's between this model and the previous
    ratioTest = random.uniform(0,1)                                # Generate a random number to compare with the ratio
    
    # If the random number is less than the ratio (which it will always be for likely models, and sometimes may be for unlikely)
    if ratioTest < ratio:
        
        # Update the model to be this new model, and update the saved best fit parameters
        chi2I = chi2F
        chi2_reduced = chi2F / (len(mu)-2)
        omsaved = om
        olsaved = ol
        w0saved = w0
        wasaved = wa
        orrsaved = orr
        
        if abs(1-chi2_reduced) < 0.15:
            omarray.append(omsaved)
            olarray.append(olsaved)
            orrarray.append(orrsaved)
            w0array.append(w0saved)
            waarray.append(wasaved)

omarray.pop(0)
olarray.pop(0)
orrarray.pop(0)
w0array.pop(0)
waarray.pop(0)

likelihoodarray.pop(0)
omlikelihoodarray.pop(0)
ollikelihoodarray.pop(0)
orrlikelihoodarray.pop(0)
w0likelihoodarray.pop(0)
walikelihoodarray.pop(0)

delta_om = 0.5*(max(omarray)-min(omarray))
delta_ol = 0.5*(max(olarray)-min(olarray))
delta_orr = 0.5*(max(orrarray)-min(orrarray))
delta_w0 = 0.5*(max(w0array)-min(w0array))
delta_wa = 0.5*(max(waarray)-min(waarray))
            
print('Best fit values are (om,ol)=(%.3f \u00B1 %.3f, %.3f \u00B1 %.3f, %.3f \u00B1 %.3f, %.3f \u00B1 %.3f, %.3f \u00B1 %.3f)'%(omsaved, delta_om, olsaved, delta_ol, orrsaved, delta_orr, w0saved, delta_w0, wasaved, delta_wa))
print('Reduced chi^2 for the best fit is %0.2f'%chi2_reduced)

# Generate the final model
emptymodel = dist_mod(zs,om=0.0,ol=0.0,w0=0,wa=0,orr=0)
model = dist_mod(zs,om=omsaved,ol=olsaved, w0=w0saved,wa=wasaved,orr=orrsaved)
modelscr = np.sum((model-mu)/muerr**2)/np.sum(1./muerr**2)
model = model-modelscr

modelMAX = dist_mod(zs,om=omsaved+delta_om,ol=olsaved+delta_ol,w0=w0saved+delta_w0,wa=wasaved+delta_wa,orr=orrsaved+delta_orr)
modelscrMAX = np.sum((model-mu)/muerr**2)/np.sum(1./muerr**2)
modelMAX = modelMAX-modelscrMAX

modelMIN = dist_mod(zs,om=omsaved-delta_om,ol=olsaved-delta_ol,w0=w0saved-delta_w0,wa=wasaved-delta_wa,orr=orrsaved-delta_orr)
modelscrMIN = np.sum((model-mu)/muerr**2)/np.sum(1./muerr**2)
modelMIN = modelMIN-modelscrMIN

# Plot it to see what it looks like, this is called a Hubble diagram
plt.errorbar(zs,mu,yerr=muerr,fmt='.',elinewidth=0.7,markersize=4,alpha=0.5)
plt.plot(zs,model,'-.',color='red',label='Model')
plt.plot(zs,modelMAX,'-.',color='black',label='Upper Error')
plt.plot(zs,modelMIN,'-.',color='black',label='Lower Error')
plt.xlim(0,1.0)
plt.xlabel('Redshift')
plt.ylabel('Distance Modulus')
plt.legend(frameon=False)
plt.show()

# Now plot a Hubble diagram relative to the empty model (i.e. subtract the empty model from all the data and models)
plt.errorbar(zs,mu-emptymodel,yerr=muerr,fmt='.',elinewidth=0.7,markersize=4,alpha=0.5)
plt.plot(zs,model-emptymodel,'-.',color='red',label='Model')
plt.plot(zs,modelMAX-emptymodel,'-.',color='black',label='Upper Error')
plt.plot(zs,modelMIN-emptymodel,'-.',color='black',label='Lower Error')
plt.axhline(y=0.0,ls=':',color='black')
plt.xlim(0,1.0)
plt.xlabel('Redshift')
plt.ylabel('Distance Modulus Normalised to (0,0,0,0,0)')
plt.legend(frameon=False)
plt.show()

fig, axs = plt.subplots(2,2)
fig.suptitle('Likelihoods')
axs[0,0].plot(omlikelihoodarray, likelihoodarray, '.')
axs[0,1].plot(ollikelihoodarray, likelihoodarray, '.')
axs[1,0].plot(orrlikelihoodarray, likelihoodarray, '.')
axs[1,1].plot(w0likelihoodarray, likelihoodarray, '.')
plt.show()

plt.plot(walikelihoodarray, likelihoodarray, '.')
```
